```python
quarter =['Q1','Q2','Q3','Q4']
```

```python
p = data["Quarter"][0]
p[0:2]
data['quarter']= 0

for i in range(42):
    p = data["Quarter"][i]
    data['quarter'][i]= p[0:2]

data.head()
```

```python
df=pd.get_dummies(data['quarter'])
df
coke=pd.concat([data,df],axis=1)
coke
```

```python
coke['T']=np.arange(1,43)
coke['T_sq']=coke['T']**2
coke['log_sales']=np.log(coke['Sales'])
coke
```

```python
Train = coke.head(30)
Test = coke.tail(10)
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
linear_model = smf.ols('Sales~T',data=Train).fit()                                                
pred_linear = pd.Series(linear_model.predict(pd.DataFrame(Test['T'])))
rmse_linear = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_linear))**2))
print("RMSE Linear: ",rmse_linear)
```

```python
Exp_model = smf.ols('log_sales~T',data=Train).fit()                                                
pred_exp = pd.Series(Exp_model.predict(pd.DataFrame(Test['T'])))
rmse_exp = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_exp)))**2))
print("RMSE EXP: ",rmse_exp)
```

```python
Quad_model = smf.ols('Sales~T+T_sq',data=Train).fit()                                              
pred_Quad = pd.Series(Quad_model.predict(Test[["T","T_sq"]]))
rmse_Quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_Quad))**2))
print("RMSE Quadratic: ",rmse_Quad)
```

```python
add_sea_model = smf.ols('Sales~Q1+Q2+Q3+Q4',data=Train).fit()                                   
pred_add_sea = pd.Series(add_sea_model.predict(Test[['Q1', 'Q2', 'Q3', 'Q4']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea))**2))
print("RMSE Additive seasonality: ",rmse_add_sea)
```

```python
add_sea_Quad_model = smf.ols('Sales~T+T_sq+Q1+Q2+Q3+Q4',data=Train).fit()       
pred_add_sea_quad = pd.Series(add_sea_Quad_model.predict(Test[['Q1', 'Q2', 'Q3', 'Q4','T','T_sq']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea_quad))**2))
print("RMSE Additive Seasonality Quadratic:",rmse_add_sea_quad )
```

```python
Mul_sea_model = smf.ols('log_sales~Q1+Q2+Q3+Q4',data = Train).fit()       
pred_Mult_sea = pd.Series(Mul_sea_model.predict(Test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_sea)))**2))
print("RMSE Multiplicative Seasonality:",rmse_Mult_sea)
```

```python
Mul_Add_sea_model = smf.ols('log_sales~T+Q1+Q2+Q3+Q4',data = Train).fit()    
pred_Mult_add_sea = pd.Series(Mul_Add_sea_model.predict(Test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_add_sea)))**2))
print("RMSE Multiplicative Additive Seasonality:",rmse_Mult_add_sea )
```

```python
Q = {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_Mult_add_sea"]),         
        "RMSE_Values":pd.Series([rmse_linear,rmse_exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea,rmse_Mult_add_sea])}              
table_rmse=pd.DataFrame(Q)
table_rmse
```
