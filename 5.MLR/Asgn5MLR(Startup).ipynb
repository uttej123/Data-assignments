```python
model=smf.ols("Profit~RDS+Admin+MS",data=data1).fit()
model.summary()
```

```python
model.params
```

```python
model.tvalues,'\n',np.round(model.pvalues,4)
```

```python
model.rsquared,model.rsquared_adj
```

```python
a=smf.ols("Profit~Admin",data=data1).fit()
a.pvalues
```

```python
m=smf.ols("Profit~MS",data=data1).fit()
m.pvalues
```

```python
am=smf.ols("Profit~Admin+MS",data=data1).fit()
np.round(am.pvalues,4)
```

```python
rsq_rds=smf.ols("RDS~Admin+MS",data=data1).fit().rsquared
vif_rds=1/(1-rsq_rds)
vif_rds
```

```python
rsq_A=smf.ols("Admin~RDS+MS",data=data1).fit().rsquared
vif_A=1/(1-rsq_A)
vif_A
```

```python
rsq_ms=smf.ols("MS~Admin+RDS",data=data1).fit().rsquared
vif_ms=1/(1-rsq_ms)
vif_ms
```

```python
q=pd.DataFrame({"Variable":["RDS","Admin","MS"],"VIF":[vif_rds,vif_A,vif_ms]})
q
```

```python
sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
model.fittedvalues
```

```python
model.resid
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "RDS", fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Admin", fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "MS", fig=fig)
plt.show()
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
c
```

```python
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(data1)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
np.argmax(c),np.max(c)
```

```python
influence_plot(model)
plt.show()
```
