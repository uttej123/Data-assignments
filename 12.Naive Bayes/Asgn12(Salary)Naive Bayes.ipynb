```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
NB=pd.read_csv('SalaryData_Train.csv')
NB
```

```python
dum=pd.get_dummies(NB.iloc[:,0:13])
dum 
```

```python
alc=LabelEncoder()
dum2=alc.fit_transform(NB['Salary'])
dum2
```

```python
kal=pd.Series(dum2)
kal
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
```

```python
model=MB()
```

```python
model.fit(dum,dum2)
```

```python
NB1=pd.read_csv('SalaryData_Test.csv')
NB1
```

```python
x=alc.fit_transform(NB1['Salary'])
```

```python
y=pd.Series(x)
y
```

```python
z=pd.get_dummies(NB1.iloc[:,0:13])
z
```

```python
pre=model.predict(z) 
```

```python
np.mean(pre==y)
```
